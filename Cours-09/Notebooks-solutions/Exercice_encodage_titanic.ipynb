```python
import pandas as pd
df = pd.read_csv("titanic.csv", sep=";")
df.head()
```

```python
#determiner le type de chaque serie
df.dtypes
```

```python
# ["Name"] contient des informations personnelles
#On constate que c'est unique pour chaque passager
len(df.name.unique() )
df['sexe_male'] = df.sex.map({'female':0, 'male':1})
df
```

```python
#Avec dummy variables
pd.get_dummies(df.sex)
```

```python
#Avec dummy variables sur tout le dataframe (serie sex)
#La serie sex est supprimée
df_dummy = pd.get_dummies(df, columns=["sex"])
df_dummy
```

```python
#Obtention de la serie (2 series et à la fin une seule serie dummy)
#Note: iloc prend toutes les lignes et la serie 1 au lieu de la serie 0
#Voir https://www.shanelynn.ie/select-pandas-dataframe-rows-and-columns-using-iloc-loc-and-ix/
df_res=pd.get_dummies(df.sex, prefix='sexe').iloc[:, 1:]
df_res
```
