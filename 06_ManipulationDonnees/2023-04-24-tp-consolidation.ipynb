```python
import pandas as pd
df = pd.read_csv("./Data/LASI_CCK_11_Twitter.csv")
df.head()
```

```python
df["mot_count"] = df["text"].map(lambda x : len(str(x).split(" ") ))
df.head()
```

```python
## string length
df["car_count"] = df["text"].map(lambda x : len(str(x) ))
df.head()
```

```python
### mean length
def moyenne_mot(phrase):
    mots = phrase.split(" ")
    return sum(len(mot) for mot in mots)/len(mots)

df["long_moy_mot"] = df["text"].map(lambda x : moyenne_mot(x))
df.head()
```

```python
### stop words
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
mots_stop = stopwords.words('english')
df["nbre_stopwords"] = df["text"].map(lambda x: len([x  for x in x.split(" ")
                                                     if x in mots_stop]))
df[["text","nbre_stopwords"]].head()
```

```python
### words beginning with @
df["nbre_speciaux"] = df["text"].map(lambda x : len([x for x in x.split(" ") 
                                                    if x.startswith("@")]))
df[["text","nbre_speciaux"]].head()
```

```python
### digits
df["nbre_digits"] = df["text"].map(lambda x : len([x for x in x.split(" ") 
                                                    if x.isdigit()]))
df[["text","nbre_digits"]].head()
```

```python
import pandas as pd
df = pd.read_csv("./Data/LASI_CCK_11_Twitter.csv")
df.head()
```

```python
from textblob import TextBlob
import nltk
nltk.download('punkt')
TextBlob(df['text'][1]).words
```

```python
### Tokenizing
df["tokens"] = df["text"].map(lambda x: TextBlob(x).words)
df[["text", "tokens"]].head()
```

```python
### Stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
df["stems"] = df["text"].map(lambda x: " ".join([ps.stem(mot) for mot 
                                                in x.split()]))
df[["text", "stems"]].head()
```

```python
### Lemming
from textblob import Word
nltk.download("wordnet")
df["lems"] = df["text"].map(lambda x: " ".join([Word(mot).lemmatize("v") 
                                                for mot 
                                                in x.split(" ")]))
df[["text", "lems"]].head()
```
